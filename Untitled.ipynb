```python
import matplotlib.pyplot as plt
import folium
```

```python
# Membuat peta
m = folium.Map(location=[df['lon'].mean(), df['lat'].mean()], zoom_start=10)

# Menambahkan marker ke peta
for i, row in df.iterrows():
    folium.CircleMarker(
        location=(row['lon'], row['lat']),
        radius=row['weight'] * 2,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
    ).add_to(m)

# Menampilkan peta
m
```

```python
count_df = df['taskStatusLabel'].value_counts()
print(count_df)
```

```python
plt.pie(count_df, labels=count_df.index, autopct='%1.1f%%')
plt.title('Task Status Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Assuming you have a DataFrame named 'df' with the data
filtered_df = df[df['taskStatusLabel'] == 'Failed']
task_detail_status_counts = filtered_df['taskDetailStatusLabel'].value_counts()

print(task_detail_status_counts)
```
